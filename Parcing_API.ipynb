```python
import requests
import json
import pandas as pd
import datetime as dt
```

```python
url = 'https://api.binance.us/api/v3/klines'
symbol = 'BTCBUSD'
interval ='1h'
start=str(int(dt.datetime(2021,1,1).timestamp()*1000))
end=str(int(dt.datetime(2021,12,1).timestamp()*1000))
```

```python
params ={'symbol':symbol,
         'interval':interval,
         'endTime':end}
```

```python
response = requests.get(url,params=params).text
```

```python
response
```

```python
json_res=json.loads(response)
```

```python
df=pd.DataFrame(json_res)
df.columns=['datetime','open', 'high','low','close','volume','close_time','qav','num_trends','taker_base_vol', 'taker_quote_vol', 'ignore']
```
